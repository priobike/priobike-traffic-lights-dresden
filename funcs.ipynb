```python
import matplotlib.pyplot as plt
import math

y = [
    min(
        1.0, 
        0.7
        + ((math.sin((math.pi / 4) * (h - 4)) + 1) / 2) * 0.1
        + ((math.sin((math.pi / 12) * (h - 6)) + 1) / 2) * 0.3
    )
    for h in range(24)
]
x = [h for h in range(24)]
plt.ylim(0, 1)
plt.plot(x, y)
```
